```python
average_rating = pd.DataFrame(ratings.groupby('ISBN')['bookRating'].mean())
average_rating['ratingCount'] = pd.DataFrame(ratings.groupby('ISBN')['bookRating'].count())
average_rating.sort_values('ratingCount', ascending=False).head()
```

```python
#print(ratings.shape)
counts1 = ratings['userID'].value_counts()
ratings = ratings[ratings['userID'].isin(counts1[counts1 >= 200].index)]
#print(ratings.shape)
counts = ratings['ISBN'].value_counts()
ratings = ratings[ratings['ISBN'].isin(counts[counts >= 100].index)]
#print(ratings.shape)

#Please explain what you think this logical error:
#1. Filtering with BookRatings after execluding userIDs wouldn't help in decreasing the data (527k),
#    On the otherhand with ISBN it reduced to (13k) only.
#2. People with less than 200 rating will mostly common to not being a typical readers so they tend to give most of the time less ratings, so execluding the
#    less popular books will not tend to decrease the size of data(as they already execluded before when execluding the low rating people).
```

```python
ratings_pivot = ratings.pivot(index='userID', columns='ISBN').bookRating
userID = ratings_pivot.index
ISBN = ratings_pivot.columns
print(ratings_pivot.shape)
ratings_pivot.head()
```

```python
bones_ratings = ratings_pivot['0316666343']
similar_to_bones = ratings_pivot.corrwith(bones_ratings)
corr_bones = pd.DataFrame(similar_to_bones, columns=['pearsonR'])
corr_bones.dropna(inplace=True)
corr_summary = corr_bones.join(average_rating['ratingCount'])
corr_summary[corr_summary['ratingCount']>=300].sort_values('pearsonR', ascending=False).head(10)

```

```python
#This Recommender System based on "The Most Similar movie" to each Top n Movies selected.
def getTopNRecommentationsforUser(userID,n):
    Search = pd.DataFrame(ratings.loc[ratings['userID'] == userID]).sort_values('bookRating', ascending=False).head(n)
    RecommendedMovies = pd.DataFrame()
    for i in Search['ISBN']:
        x = ratings_pivot[i]
        Similar = ratings_pivot.corrwith(x)
        DF = pd.DataFrame(Similar, columns=['Similarity']).sort_values('Similarity', ascending=False).head(2)
        recommendations = DF[1:2]
        RecommendedMovies = RecommendedMovies.append(recommendations)
    return print(RecommendedMovies)
```
